Preview the first few rows of each of the four main sheets to confirm structure and locate REMEDY_02 sections.

## 5. Report Row Counts for All Four Sheets

Count and print the exact number of rows in each of the four main sheets after REMEDY_02 population.

```python
# Count and print the exact number of rows (including header) in each main sheet
def count_rows(sheet_name):
    ws = wb[sheet_name]
    return ws.max_row

row_counts = {}
for sheet in ['plan_master', 'plan_items_master', 'plan_item_options', 'plan_comparison_matrix']:
    row_counts[sheet] = count_rows(sheet)
    print(f"{sheet}: {row_counts[sheet]} rows (including header)")

row_counts
```
